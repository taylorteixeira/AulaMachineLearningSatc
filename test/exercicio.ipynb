```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Dados das questões
datasets = {
    "Diagnóstico Médico": {"TP": 150, "TN": 800, "FP": 30, "FN": 20}
}

# Função para calcular métricas
def calcular_metricas(TP, TN, FP, FN):
    y_true = np.array([1]*TP + [0]*TN + [1]*FN + [0]*FP)
    y_pred = np.array([1]*(TP+FP) + [0]*(TN+FN))
    
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    cm = confusion_matrix(y_true, y_pred)
    
    return acc, prec, rec, f1, cm

# Resultados
resultados = {}
for nome, valores in datasets.items():
    acc, prec, rec, f1, cm = calcular_metricas(valores['TP'], valores['TN'], valores['FP'], valores['FN'])
    resultados[nome] = {
        "Acurácia": acc,
        "Precisão": prec,
        "Recall": rec,
        "F1-Score": f1,
        "Matriz de Confusão": cm
    }

resultados
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.datasets import load_iris

# Carregar o dataset e dividir em treino e teste
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

# Treinamento do modelo
modelo_classificacao = DecisionTreeClassifier()
modelo_classificacao.fit(X_train, y_train)

# Previsões e avaliação
y_pred = modelo_classificacao.predict(X_test)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='weighted')
rec = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
cm = confusion_matrix(y_test, y_pred)

{"Acurácia": acc, "Precisão": prec, "Recall": rec, "F1-Score": f1, "Matriz de Confusão": cm}
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.datasets import load_boston

# Carregar o dataset e dividir em treino e teste
data = load_boston()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

# Treinamento do modelo de regressão
modelo_regressao = LinearRegression()
modelo_regressao.fit(X_train, y_train)

# Previsões e avaliação
y_pred = modelo_regressao.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

{"MSE": mse, "MAE": mae, "R²": r2}
```
